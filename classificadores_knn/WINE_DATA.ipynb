```python
import pandas as pd
import numpy as np
```

```python
from sklearn.datasets import load_wine
wine=load_wine()


data=pd.DataFrame(data=np.c_[wine['data'],wine['target']],columns=wine['feature_names']+['target'])
wine_5_feat_input = data[['malic_acid' ,'alcalinity_of_ash','ash', 'nonflavanoid_phenols','color_intensity', 'target']].copy()
```

```python
wine_input = data.loc[:,data.columns != 'target'].values
wine_target = data.loc[:,['target']].values
target_names = ['Tipo 1', 'Tipo 2', 'Tipo 3']
```

```python
import sys
from time import sleep
from tqdm.notebook import tqdm_notebook
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestCentroid
from sklearn.metrics import make_scorer, confusion_matrix, accuracy_score
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix

it = 0
sc = StandardScaler()
```

```python
from sklearn.neighbors import KNeighborsClassifier
dmc = NearestCentroid(metric='euclidean',)
nn = KNeighborsClassifier(n_neighbors=1, metric='euclidean')
knn3 = KNeighborsClassifier(n_neighbors=3, metric='euclidean')
knn5 = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
knn7 = KNeighborsClassifier(n_neighbors=7, metric='euclidean')
```

```python
acc_test_result = []
confusion_matrix_list = []
acc_test_result_nn = []
confusion_matrix_list_nn = []
acc_test_result_knn3 = []
confusion_matrix_list_knn3 = []
acc_test_result_knn5 = []
confusion_matrix_list_knn5 = []
acc_test_result_knn7 = []
confusion_matrix_list_knn7 = []
acc_test_result_5 = []
confusion_matrix_list_5 = []
acc_test_result_nn_5 = []
confusion_matrix_list_nn_5 = []
acc_test_result_knn3_5 = []
confusion_matrix_list_knn3_5 = []
acc_test_result_knn5_5 = []
confusion_matrix_list_knn5_5 = []
acc_test_result_knn7_5 = []
confusion_matrix_list_knn7_5 = []

for it in tqdm_notebook(range(50)):
    X_train, X_test, X_train_5, X_test_5, y_train, y_test = train_test_split(wine_input, wine_5_feat_input, wine_target , test_size=0.20, stratify=wine_target)
    #X_train = sc.fit_transform(X_train)
    #X_test = sc.transform(X_test)
    #X_train_5 = sc.fit_transform(X_train)
    #X_test_5 = sc.transform(X_test)

    dmc.fit(X_train, y_train.ravel())
    y_pred_wine = dmc.predict(X_test)
    acc_test_wine =  accuracy_score(y_test, y_pred_wine)
    acc_test_result.append(acc_test_wine)
    confusion_matrix_list.append(confusion_matrix(y_test, y_pred_wine))
    nn.fit(X_train, y_train.ravel())
    y_pred_wine_nn = nn.predict(X_test)
    acc_test_wine_nn =  accuracy_score(y_test, y_pred_wine_nn)
    acc_test_result_nn.append(acc_test_wine_nn)
    confusion_matrix_list_nn.append(confusion_matrix(y_test, y_pred_wine_nn))
    knn3.fit(X_train, y_train.ravel())
    y_pred_wine_knn3 = knn3.predict(X_test)
    acc_test_wine_knn3 =  accuracy_score(y_test, y_pred_wine_knn3)
    acc_test_result_knn3.append(acc_test_wine_knn3)
    confusion_matrix_list_knn3.append(confusion_matrix(y_test, y_pred_wine_knn3))
    knn5.fit(X_train, y_train.ravel())
    y_pred_wine_knn5 = knn5.predict(X_test)
    acc_test_wine_knn5 =  accuracy_score(y_test, y_pred_wine_knn5)
    acc_test_result_knn5.append(acc_test_wine_knn5)
    confusion_matrix_list_knn5.append(confusion_matrix(y_test, y_pred_wine_knn5))
    knn7.fit(X_train, y_train.ravel())
    y_pred_wine_knn7 = knn7.predict(X_test)
    acc_test_wine_knn7 =  accuracy_score(y_test, y_pred_wine_knn7)
    acc_test_result_knn7.append(acc_test_wine_knn7)
    confusion_matrix_list_knn7.append(confusion_matrix(y_test, y_pred_wine_knn7))

    dmc.fit(X_train_5, y_train.ravel())
    y_pred_wine_5 = dmc.predict(X_test_5)
    acc_test_wine_5 =  accuracy_score(y_test, y_pred_wine_5)
    acc_test_result_5.append(acc_test_wine_5)
    confusion_matrix_list_5.append(confusion_matrix(y_test, y_pred_wine_5))
    nn.fit(X_train_5, y_train.ravel())
    y_pred_wine_nn_5 = nn.predict(X_test_5)
    acc_test_wine_nn_5 =  accuracy_score(y_test, y_pred_wine_nn_5)
    acc_test_result_nn_5.append(acc_test_wine_nn_5)
    confusion_matrix_list_nn_5.append(confusion_matrix(y_test, y_pred_wine_nn_5))
    knn3.fit(X_train_5, y_train.ravel())
    y_pred_wine_knn3_5 = knn3.predict(X_test_5)
    acc_test_wine_knn3_5 =  accuracy_score(y_test, y_pred_wine_knn3_5)
    acc_test_result_knn3_5.append(acc_test_wine_knn3_5)
    confusion_matrix_list_knn3_5.append(confusion_matrix(y_test, y_pred_wine_knn3_5))
    knn5.fit(X_train_5, y_train.ravel())
    y_pred_wine_knn5_5 = knn5.predict(X_test_5)
    acc_test_wine_knn5_5 =  accuracy_score(y_test, y_pred_wine_knn5_5)
    acc_test_result_knn5_5.append(acc_test_wine_knn5_5)
    confusion_matrix_list_knn5_5.append(confusion_matrix(y_test, y_pred_wine_knn5_5))
    knn7.fit(X_train_5, y_train.ravel())
    y_pred_wine_knn7_5 = knn7.predict(X_test_5)
    acc_test_wine_knn7_5 =  accuracy_score(y_test, y_pred_wine_knn7_5)
    acc_test_result_knn7_5.append(acc_test_wine_knn7_5)
    confusion_matrix_list_knn7_5.append(confusion_matrix(y_test, y_pred_wine_knn7_5))


    sleep(0.1)
```
